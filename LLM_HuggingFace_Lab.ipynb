```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch
```

1. Model Swap & Comparison




```python
base_model = "gpt2"
alt_model = "distilgpt2"  # Change if GPU allows (e.g., "Qwen/Qwen2-0.5B-Instruct")

# Load base pipeline
base_pipe = pipeline("text-generation", model=base_model)
# Load alternative pipeline
alt_pipe = pipeline("text-generation", model=alt_model)

prompt = "Explain what a Knowledge Graph is in healthcare, in 3 concise sentences."

print("=== Base Model Output ===")
print(base_pipe(prompt, max_length=80, num_return_sequences=1)[0]['generated_text'])
print("\n=== Alternative Model Output ===")
print(alt_pipe(prompt, max_length=80, num_return_sequences=1)[0]['generated_text'])
```

 2. Decoding Parameters – Explain in Your Own Words


```python
prompt = "Artificial intelligence in education will"

print("=== Greedy Decoding ===")
print(base_pipe(prompt, max_length=40, do_sample=False)[0]['generated_text'])

print("\n=== Sampling with temperature=0.2 (low randomness) ===")
print(base_pipe(prompt, max_length=40, do_sample=True, temperature=0.2)[0]['generated_text'])

print("\n=== Sampling with temperature=1.0 (high randomness) ===")
print(base_pipe(prompt, max_length=40, do_sample=True, temperature=1.0)[0]['generated_text'])

print("\n=== Top-k=20 ===")
print(base_pipe(prompt, max_length=40, do_sample=True, top_k=20)[0]['generated_text'])

print("\n=== Top-p=0.8 ===")
print(base_pipe(prompt, max_length=40, do_sample=True, top_p=0.8)[0]['generated_text'])
```
